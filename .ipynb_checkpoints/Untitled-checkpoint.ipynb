```python
for i in players_22.columns.tolist():
    if '_url' in i:
        players_22.drop(i, axis = 1, inplace= True)
    if '_id' in i:
        players_22.drop(i, axis = 1, inplace = True)
```

```python
#Take out positions from the dataframe because judging the players by how good they perform at positions they perform well at can lead to biased desicions.
positions= ['gk', 'rb', 'rcb', 'cb', 'lcb', 'lb', 'rwb', 'rdm', 'cdm', 'ldm', 'lwb', 'rm', 'rcm', 'cm', 'lcm', 'lm', 'ram', 'cam', 'lam', 'rw', 'rf', 'cf', 'lf', 'lw', 'rs', 'st', 'ls']
players_22.drop(positions, axis = 1, inplace= True)
```

```python
# More columns to drop with reasons
# 'short_name' : There is a long name which can act as an identifier
# 'club_position': The player's position in the club is less informative than performance metrics.
# 'club_jersey_number': Jersey numbers are not indicative of player skill or performance.
# 'club_joined': The date the player joined the club is less relevant compared to their performance metrics.
# 'club_contract_valid_until': Contract length is less directly tied to skill.
# 'real_face': Whether the player's face is realistic in the game is irrelevant to performance.
# Goalkeeping specific attributes are dropped since we are focusing on general player attributes:
# 'goalkeeping_diving', 'goalkeeping_handling', 'goalkeeping_kicking', 'goalkeeping_positioning', 'goalkeeping_reflexes'
columns_to_drop = [
    'short_name','club_position', 'club_jersey_number', 'club_joined', 
    'club_contract_valid_until', 'real_face', 
    'goalkeeping_diving', 'goalkeeping_handling', 'goalkeeping_kicking', 
    'goalkeeping_positioning', 'goalkeeping_reflexes', 'club_name', 'league_name'
]

# Drop the less relevant columns from the DataFrame
players_22 = players_22.drop(columns=columns_to_drop)
```

```python
players_22.head()
```

```python
#Remove columns withe 30% or more null values too.
L= []
L_less = []
for i in players_22.columns:
    if((players_22[i].isnull().sum())<(0.3*(players_22.shape[0]))):
        L.append(i)
    else:
        L_less.append(i)
players_22 = players_22[L]
```

```python
players_22
```

```python
players_names = pd.DataFrame(players_22['long_name'])
players_names
players_22.drop('long_name', axis = 1, inplace = True)
```

```python
categorical_players = players_22.select_dtypes(exclude='number')
numerical_players = players_22.select_dtypes(include='number')
```

```python
categorical_players.columns
```

```python
#Fill the null values in the numerical columns with the mean values
sc = SimpleImputer(strategy = 'mean')
scaled = sc.fit_transform(numerical_players)
numerical_players = pd.DataFrame(scaled, columns = numerical_players.columns)
```

```python
#Fill the null values in the catergorical columns with the most frequent
sc = SimpleImputer(strategy = 'most_frequent')
scaled = sc.fit_transform(categorical_players)
categorical_players = pd.DataFrame(scaled, columns = categorical_players.columns)
```

```python
#drop dob from categorical column since age already exists
categorical_players.drop(columns = ['dob', 'nationality_name', 'player_positions'], axis =1, inplace= True)
```

```python
#encoder = LabelEncoder()
#encoded_categories = categorical_players.apply(encoder.fit_transform)
#encoded_categories
```

```python
encoded_categories = pd.get_dummies(categorical_players).astype(int)
encoded_categories
```

```python
corr_matrix = numerical_players.corr()
```

```python
correlations = corr_matrix["overall"].sort_values(ascending=False)
```

```python
numeric_chosen = correlations[abs(correlations.values) >= 0.5]
numeric_chosen
```

```python

# Get the columns of numeric_chosen
chosen_columns = []
for variable, correlation in numeric_chosen.items():
    chosen_columns.append(variable)
# Create a new DataFrame with these chosen columns
numerical_players = numerical_players[chosen_columns]
```

```python
new_players = pd.concat([numerical_players, encoded_categories], axis=1).reset_index(drop = True)
```

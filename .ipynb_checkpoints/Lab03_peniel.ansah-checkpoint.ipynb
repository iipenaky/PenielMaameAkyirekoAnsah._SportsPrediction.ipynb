```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer, make_column_selector
```

```python
numerical_columns = players_21.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_columns = players_21.select_dtypes(include= 'object').columns.tolist()
# Numerical pipeline
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler()),
])

# Categorical pipeline with imputer
cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="most_frequent")),  # Fill missing values with the most frequent value
    ('onehot', OneHotEncoder())
])

# Full pipeline combining both numerical and categorical pipelines
full_pipeline = ColumnTransformer([('quant', num_pipeline, make_column_selector(dtype_include= [float64, int64])), 
('cat', cat_pipeline, make_column_selector(dtype_include = object))
                              ])
#numerical_columns
piped21 = full_pipeline.fit_transform(players_21)
piped22 = full_pipeline.fit_transform(players_22)

#prepared_columns = numerical_columns + list(full_pipeline.named_transformers_['cat']['onehot'].get_feature_names_out(categorical_columns))
#prepared_df = pd.DataFrame(piped21, columns=prepared_columns)


# Convert to DataFrame for better readability
#prepared_df21 = pd.DataFrame(piped21, columns=numerical_columns + full_pipeline.named_transformers_['cat']['onehot'].get_feature_names_out(categorical_columns).tolist())
```
